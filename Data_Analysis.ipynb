```python
vendor_sales_summary = pd.read_sql("""
WITH FreightSummary AS (
    SELECT VendorNumber,
           SUM(Freight) AS FreightCost
    FROM vendor_invoice
    GROUP BY VendorNumber
),
PurchaseSummary AS (
    SELECT  p.Description,
    p.VendorName,
           p.VendorNumber,
           p.PurchasePrice,
           p.Brand,
           pp.Volume,
           pp.Price AS ActualPrice,
           SUM(p.Quantity) AS TotalQuantity,
           SUM(p.Dollars) AS TotalSales
    FROM purchases AS p
    JOIN purchase_prices AS pp
      ON p.Brand = pp.Brand
    GROUP BY  p.Description, p.VendorName, p.VendorNumber, p.PurchasePrice, p.Brand, pp.Volume, pp.Price
    
    HAVING NOT (p.PurchasePrice = 0 AND SUM(p.Quantity) = 2015)
),
SalesSummary AS (
    SELECT VendorNo,
           Brand,
           SUM(SalesQuantity) AS TotalSalesQuantity,
           SUM(SalesDollars) AS TotalSalesDollars,
           SUM(SalesPrice) AS TotalSalesPrice,
           SUM(ExciseTax) AS TotalExciseTax
    FROM sales_
    GROUP BY VendorNo, Brand
)
SELECT ps.VendorNumber,
       ps.VendorName,
       ps.Brand,
       ps.Description,
       ps.PurchasePrice,
       ps.ActualPrice,
       ps.Volume,
       ps.TotalQuantity AS TotalPurchaseQuantity,
       ps.TotalSales AS TotalPurchaseDollars,
       ss.TotalSalesQuantity,
       ss.TotalSalesDollars,
       ss.TotalSalesPrice,
       ss.TotalExciseTax,
       fs.FreightCost
FROM PurchaseSummary AS ps
LEFT JOIN SalesSummary AS ss
    ON ps.VendorNumber = ss.VendorNo
    AND ps.Brand = ss.Brand
LEFT JOIN FreightSummary AS fs
    ON ps.VendorNumber = fs.VendorNumber
ORDER BY ps.TotalSales DESC
""", engine)
```

```python
import numpy as np

numeric_cols = [
    "PurchasePrice", "ActualPrice", "TotalPurchaseQuantity", "TotalPurchaseDollars",
    "TotalSalesQuantity", "TotalSalesDollars", "TotalSalesPrice", 
    "TotalExciseTax", "FreightCost", "GrossProfit", "ProfitMargin"
]

# Identify rows with invalid numeric values
for col in numeric_cols:
    invalid_rows = vendor_sales_summary[~vendor_sales_summary[col].apply(lambda x: np.isfinite(x))]
    if not invalid_rows.empty:
        print(f"Invalid values in {col}:")
        print(invalid_rows[[col]])
```

```python
vendor_sales_summary[numeric_cols] = vendor_sales_summary[numeric_cols].replace([np.inf, -np.inf], 0)
vendor_sales_summary[numeric_cols] = vendor_sales_summary[numeric_cols].fillna(0)
```

```python
vendor_sales_summary[numeric_cols] = vendor_sales_summary[numeric_cols].round(2)
```

```python
vendor_sales_summary.to_sql(
    "vendor_sales_summary",
    con=engine,
    if_exists="append",   # or "replace"
    index=False
)
```
